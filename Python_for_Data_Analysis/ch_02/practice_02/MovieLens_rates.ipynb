```python
import pandas as pd

upath = os.path.expanduser('data/users.dat')
rpath = os.path.expanduser('data/ratings.dat')
mpath = os.path.expanduser('data/movies.dat')


unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
mnames = ['movie_id', 'title', 'genres']


users = pd.read_csv(upath, sep='::', header=None, names=unames, encoding='latin1', engine='python')
ratings = pd.read_csv(rpath, sep='::', header=None, names=rnames, encoding='latin1', engine='python')
movies = pd.read_csv(mpath, sep='::', header=None, names=mnames, encoding='latin1', engine='python')
```

나이와 성별에 따른 어떤 영화의 평균 평점을 계산한다고 하자.
이때 모든 데이터를 하나의 테이블로 병합해서 계산하면 쉽게 처리할 수 있다.

pandas의 merge함수를 이용해서 ratings 테이블과 users 데이블을 병합하고, 그 결과를 다시 movies 테이블과 별합한다.
pandas에서는 병합하려는 두 테이블에서 중복되는 열의 이름을 킬로 사용한다.

```python
data = pd.merge(pd.merge(ratings, users), movies)
data.head(10)
```

```python
data.iloc[0]
```

```python
data['user_id']
```

성별에 따른 각 영화의 평균 평점을 구하려면 pivot_table 메서드를 사용하면 된다.

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings
```

위의 pivot_table 메서드를 사용해서 매 행마다 성별에 따른 평균 영화 평점을 담고 있는 Data Frame 객체가 생성됐다.
만일 이중 250건 이상의 평점 정보가 있는 영화의 목록만을 추려보고자 한다면
데이터를 영화제목으로 그룹화 한 후, size() 메서드를 사용해서 제목별 평점 정보 건수를 Series 객체로 얻어내서 활용하면 된다.

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
mean_ratings.columns
```

남녁 간에 호불호가 갈리는 영화를 찾아보고자 하면, 다음과 같이 하면된다.

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
mean_ratings
```

```python
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff
```

diff 컬럼에서 가장 큰 차이를 보이는 순서대로 정렬하면, 여성이 더 선호하는 영화 순서로 정렬되고,
그 역순으로 정렬하면 남성에게 선호도가 높은 영화 순서로 정렬된 Data Frame 객체를 반환받을 수 있다.

그럼 이번에는 성별에 관계없이 영화에 대한 호불호가 극명하게 나뉘는 영화를 찾아보자
호불호는 평점의 분산이나 표준편차를 통해 알아 볼 수 있다.

```python
# 평점의 표준편차
rating_std_by_title = data.groupby('title')['rating'].std()

#active_titles만 선택
rating_std_by_title = rating_std_by_title.loc[active_titles]

# 내림차순으로 정렬
rating_std_by_title.sort_values(ascending=False)[:10]
```
