```python
import mglearn
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')

%matplotlib inline
```

```python
import os
import pandas as pd

ram_prices = pd.read_csv(os.path.join(mglearn.datasets.DATA_PATH, 'ram_price.csv'))

plt.yticks(fontname='Arial')
plt.semilogy(ram_prices.date, ram_prices.price)
plt.xlabel('year')
plt.ylabel('price')
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression

# 2000년 이전을 훈련데이터로 2000년 이후 데이터를 테스트 데이터로 만든다.
data_train = ram_prices[ram_prices.date < 2000]
data_test = ram_prices[ram_prices.date >= 2000]


# 가격 예측을 위해서 날짜 특성만을 사용한다.
X_train = data_train.date[:, np.newaxis]
# 데이터와 타겟 사이의 관계를 간단하게 하기위해 로그 스케일로 변환해준다.
y_train = np.log(data_train.price)

dt_reg = DecisionTreeRegressor().fit(X_train, y_train)
l_reg = LinearRegression().fit(X_train, y_train)

# 예측은 전체 기간에 대해서 수행한다.
X_all = ram_prices.date[:, np.newaxis]

pred_tree = dt_reg.predict(X_all)
pred_lr = l_reg.predict(X_all)


# 예측값의 스케일을 되돌린다
price_tree = np.exp(pred_tree)
price_lr = np.exp(pred_lr)
```

```python
plt.figure(figsize=(12, 12))
plt.semilogy(data_train.date, data_train.price, label='train dataset')
plt.semilogy(data_test.date, data_test.price, label='test dataset')
plt.semilogy(ram_prices.date, price_tree, label='DTR prediction values')
plt.semilogy(ram_prices.date, price_lr, label='lr prediction values')
plt.legend()
```

두 모델은 확연한 차이를 보여준다. 선형모델은 우리가 아는대로 직선으로 데이터를 근사해냈지만, 트리 모델은 훈련데이터를 완벽하게 예측한다.
트리 복잡도에 제한을 두지 않아서 전체 데이터셋을 모두 기억하기 때문이다. 하지만 모델이 가진 데이터 범위 밖으로 나가면 단순히
마지막 포인트를 이용해서 예측하는것이 전부이다. 트리모델은 훈련데이터 밖의 새로운 데이터를 예측할 능력이 없다.
이는 모든 트리기반 모델의 공통된 단점이다.
