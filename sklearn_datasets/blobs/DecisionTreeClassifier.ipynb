결정트리는 스무고개와 동일한 원리로 작동하는 알고리즘이며 각각의 질문은 if / else 분기와 같으며 조건 값은 텍스트가 아닌 벡터화된 실수값이다.
결정트리는 분류에서 뿐만아니라 회귀에서도 사용되는 모델이다.

결정트리에서 계속해서 노드 분기를 통해 분류를 진행하다보면 알고리즘 작동 원리에 의거해서 과대적합이 될 수 밖에 없다.
과대적합은 언제나 일반화를 방해하고 실제 분류를 방해하는 요소이기 때문에 과대적합을 적절히 방지해줘야 한다.

결정트리 모델에서 과대적합을 방지하는 방법은 트리 생성을 일찍 중단하는 전략 (사전 가지치기)과 트리를 만든 후게 데이터 포인트가 적은 노드를 삭제하거나 병합하는 전략이다.

사전 가지치기 방식은 트리의 ㄱ최대 깊이나 리프의 최대 개수를 제한하거나 또는 노드가 분할하기 위한 포인트의 최소 개수를 지정하는 것이다. 미리 분기상의 조건들을 통해 특정 조건 값에 해당하면 더이상의 분기를 하지 않도록 유도하는 것.

사이킷 런에서는 사전 가지치기 방식만을 제공한다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')

%matplotlib inline

from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=42)

dt_clf = DecisionTreeClassifier(random_state=0)
dt_clf.fit(X_train, y_train)

print('train dataset accuracy : {:.3f}'.format(dt_clf.score(X_train, y_train)))
print('test dataset accuracy : {:.3f}'.format(dt_clf.score(X_test, y_test)))
```

학습 세트의 정확도는 100이다. 즉 트리는 훈련 데이터의 모든 레이블을 완벽하게 기억할 만큼 충분히 깊게 만들어졌다.
과대 적합되다 보니 테스트 세트에서의 정확도는 훈련 데이터에서의 정확도보다는 조금 떨어진다.

결정트리의 깊이를 제한하지 않으면 트리는 무한정 깊어지고 복잡해질 수 있다.
그래서 가지치기를 하지 않은 트리는 과대적합되기 쉽고 새로운 데이터에 잘 일반화되지 않는다. 
사전 가지치기를 위한 옵션은 max_depth 속성값을 통해 부여 할 수 있다.

가지치기를 통해 train set의 정확도는 다소 낮아 질 수는 있어도 test set의 정확도는 높아진다.

```python
dt_clf = DecisionTreeClassifier(max_depth=4, random_state=0)
dt_clf.fit(X_train, y_train)

print('train dataset accuracy : {:.3f}'.format(dt_clf.score(X_train, y_train)))
print('test dataset accuracy : {:.3f}'.format(dt_clf.score(X_test, y_test)))
```

트리모듈의 export_graphviz 람수를 이용하면 트리를 시각화해서 볼 수도 있다.

```python
from sklearn.tree import export_graphviz

export_graphviz(dt_clf, out_file='dt.dot', class_names=['malicious', 'benign'], feature_names= cancer.feature_names, impurity=False, filled=True)
```

```python
import graphviz
from IPython.display import display

with open('dt.dot') as f:
    dot_graph = f.read()
display(graphviz.Source(dot_graph))
```

전체 트리를 살펴보는 것은 어려울 수 있으니, 대신 트리가 어떻게 작동하는지 요약하는 속성들을 사용할 수 있다. 
가장 널리 사용되는 속성은 트리를 만드는 결정에 각 특성이 얼마나 중요한지를 평가하는 특성 중요도 (feature importance)이다.
이 값은 0과 1 사이의 숫자로, 각 특성에 대해 0은 전혀 사용되지 않았다는 뜻이고 1은 완벽하게 타깃클래스를 예측했다는 의미이다.
특성 중요도의 전체합은 1이다.

```python
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.figure(figsize=(15, 15))
    plt.barh(np.arange(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel('feature importances')
    plt.ylabel('features names')
    plt.ylim(-1, n_features)


    
plot_feature_importances_cancer(dt_clf)
```
