데이터를 시각화하는 좋은 방법중 하나는 산점도 그래프를 활용하는 것이다. 산점도는 데이터에서 한 특성을 x축에 놓고, 다른 특성을 y축에 위치시켜서 각 데이터 포인트를 하나의 점으로 나타내는 그래프이다.
데이터의 특성이 많지 않은 경우에 활용하기 좋은 방법이다.

```python
pd.plotting.scatter_matrix(iris_df.iloc[:, :-1], c=iris_df.iloc[:, -1], figsize=(15, 15), hist_kwds={'bins':20}, marker='o', cmap='plasma')
```

데이터에 대한 기본적인 분석이 끝나면 데이터를 훈련용 테스트용으로 분할 후, 적용가능한 여러가지의 모델을 적용해서
각각의 정확도를 기반으로 모델의 성능을 측정해야한다.

sklearn의 모든 머신러닝 모델은 Estimator라는 파이썬 클래스로 구현되어 있다. 
모델을 사용하기 위해서는 클래스로부터 객체를 우선 생성해야한다.
이때 모델에 필요한 매개변수를 각각 부여한다.

```python
from sklearn.model_selection import train_test_split


X = iris_df.iloc[:, :-1]
y = iris_df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(y_train.value_counts())
print(y_test.value_counts())
```

```python
type(X_train)
```

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

# 훈련 데이터셋으로부터 모델을 만드려면 knn 객체의 fit메서드를 사용해야한다.
# 이 메서드는 훈련 데이터인 Numpy 배열 X_train과 이에 상응하는 레이블 배열인 y_train을 매개변수로 받는다

knn.fit(X_train, y_train)
```

fit메서드는 knn 객체 자체를 반환한다. 그리고 knn 객체 자체를 변경시킨다. 그래서 knn 객체가 문자열 형태로 출력된다.

```python
y_pred = knn.predict(X_test)

print('테스트 세트의 정확도 : {:.2f}'.format(np.mean(y_pred==y_test)))
```

```python
print('테스트 세트의 정확도 : {:.2f}'.format(knn.score(X_test, y_test)))

# 이 방법을 사용할 때는 별도의 변수를 선언해서 pred 저장할 필요성이 줄어든다. 간결성 up
```
