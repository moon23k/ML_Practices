```python
import pandas as pd

url = 'https://www.seoul.go.kr/coronaV/coronaStatus.do'

table = pd.read_html(url)
table
```

```python
type(table)
```

```python
len(table)
```

```python
table[3]
```

```python
df = table[3]
df.shape
```

```python
last_day = df.loc[0, '확진일']
last_day
```

```python
last_day = last_day.replace('.', '')
last_day
```

```python
file_path = 'data/'
file_name = f'seoul-covid19_{last_day}.csv'

file_path + file_name
```

```python
df.to_csv(file_path + file_name)
```

```python
# 제대로 저장되었는지 확인

pd.read_csv(file_path + file_name)
```
